# Pakistani Politician Image Classification - EDA

Exploratory Data Analysis for the dataset

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from PIL import Image
import numpy as np

sns.set_style('whitegrid')
```

## 1. Dataset Statistics

```python
# Count images per class
def count_images(root_dir):
    counts = {}
    for class_name in os.listdir(root_dir):
        class_path = Path(root_dir) / class_name
        if class_path.is_dir():
            images = list(class_path.glob('*.jpg')) + list(class_path.glob('*.png'))
            counts[class_name] = len(images)
    return counts

train_counts = count_images('dataset/train')
val_counts = count_images('dataset/val')
test_counts = count_images('dataset/test')

print(f"Train: {sum(train_counts.values())} images")
print(f"Val: {sum(val_counts.values())} images")
print(f"Test: {sum(test_counts.values())} images")
```

## 2. Class Distribution

```python
# Plot class distribution
plt.figure(figsize=(14, 6))
plt.bar(train_counts.keys(), train_counts.values())
plt.xlabel('Politician')
plt.ylabel('Number of Images')
plt.title('Training Set - Class Distribution')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 3. Sample Images

```python
# Display sample images from each class
fig, axes = plt.subplots(4, 4, figsize=(16, 16))
axes = axes.flatten()

for idx, class_name in enumerate(sorted(train_counts.keys())):
    class_path = Path('dataset/train') / class_name
    images = list(class_path.glob('*.jpg'))[:1]
    
    if images:
        img = Image.open(images[0])
        axes[idx].imshow(img)
        axes[idx].set_title(class_name)
        axes[idx].axis('off')

plt.tight_layout()
plt.show()
```
